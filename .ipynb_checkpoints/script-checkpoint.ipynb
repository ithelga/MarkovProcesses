### Требуется определить:

### 1) вероятность того, что за 10 шагов система перейдет из состояния 4 в состояние 6;

```python
def Pk(P, k):
    Z = P.copy()
    for i in range(k - 1):
        Z = np.dot(P, Z)
    return Z

st1 = 4
st2 = 6
k = 10

print(f'Вероятность, что за 10 шагов система перейдет из состояния 4 в состояние 6: {Pk(P, k)[st1-1][st2-1]}')
```

### 2) вероятности состояний системы спустя 8 шагов, если в начальный момент вероятность состояний были следующими A= (0,09;0,08;0,09;0,13;0,08;0,08;0,11;0,07;0,05;0,08;0,14)

```python
def Pa(matrix, k, A0):
    AN = np.dot(A0, Pk(matrix, k))
    return AN

k = 8
A0 = np.array([0.09, 0.08, 0.09, 0.13, 0.08, 0.08, 0.11, 0.07, 0.05, 0.08, 0.14])

print(f'Вероятности состояний системы спустя 8 шагов:')
print(Pa(P, k, A0))
```

### 4) вероятность перехода из состояния 2 в состояние 10 не позднее чем за 6 шагов

```python
def p_until_k(P, k):
    P_start = np.copy(P)
    P_end = np.copy(P)
    for t in range(k):
        P_start = skip_dot(P, P_start, t)
        P_end += P_start
    return P_end

st1 = 2
st2 = 10
k = 6
print(f'вероятность перехода из состояния 2 в состояние 10 не позднее чем за 6 шагов: {p_until_k(P, k)[st1-1][st2-1]}')
```

### 5) среднее количество шагов для перехода из состояния 6 в состояние 3

```python
def n_avg_step(P):
    P_start = np.copy(P)
    P_end = np.copy(P)
    for t in range(1,1000):
        P_start = skip_dot(P, P_start,t)
        P_end += t * P_start
    return P_end

st1 = 6
st2 = 3
print(f'Cреднее количество шагов для перехода из состояния 6 в состояние 3: {round(n_avg_step(P)[st1-1][st2-1], 3)}')
```

### 6) вероятность первого возвращения в состояние 10 за 6 шагов

```python
from functools import lru_cache

def p_first_return(P, k):
    Pc = np.copy(P)
    
    @lru_cache(maxsize=128)
    def Fjj(k):
        u = np.zeros((size, size))
        for m in range(1, k):
            u += Fjj(m) * Pk(Pc, k - m)
        u = Pk(Pc, k) - u
        return u
    
    return np.diagonal(Fjj(k))

k = 6
st = 10
p_first_return = p_first_return(P,k)
print(f'вероятность первого возвращения в состояние 10 за 6 шагов: {p_first_return[st-1]}')
```

### 7) вероятность возвращения в состояние 11 не позднее чем за 9 шагов

```python
def p_return_until_k(P, k):
    Pc = np.copy(P)
    list_u = []

    @lru_cache(maxsize=128)
    def Fjj(k):
        u = np.zeros((size, size))
        for m in range(1, k):
            u+= Fjj(m) * Pk(Pc, k - m)
        u = Pk(Pc, k) - u
        list_u.append(np.diagonal(u))
        return u
    Fjj(k)
    return sum(list_u)

k = 9
st = 11
p_return_until_k = p_return_until_k(P, k)
print(f'вероятность возвращения в состояние 11 не позднее чем за 9 шагов: {p_return_until_k[st-1]}')
```

### 8) среднее время возвращения в состояние 8

```python
def T_avg_return(P):
    Pc = np.copy(P)
    list_u = []

    @lru_cache(maxsize=128)
    def Fjj(k):
        u = 0
        for m in range(1,k):
            u += Fjj(m) * Pk(Pc, k - m)
        u = Pk(Pc, k) - u
        list_u.append(k * np.diagonal(u))
        return u

    Fjj(100)
    return sum(list_u)

st = 8
T_avg_return = T_avg_return(P)
print(f'среднее время возвращения в состояние 8: {T_avg_return[st-1]}')
```

## Задание 2

Задана система массового обслуживания со следующими характеристиками:
    интенсивность поступления		λ=16
    каналов обслуживания			m=6
	интенсивность обслуживания		μ=4
	максимальный размер очереди	n=14
Изначально требований в системе нет.

```python
import numpy as np

la = 16
m = 6
ny = 4
n = 14
size = m+n+1
```

### Составляем матрицу интенсивности переходов

```python
def findL(size,la,ny,m):
    L = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            if i + 1 == j:
                L[i, j] = la
            if j + 1 == i:
                if i <= m:
                    L[i, j] = i * ny
                else:
                    L[i, j] = m * ny
    return L

L = findL(size,la,ny,m)
print(L)
```

### a) Граф марковского процесса

### a) система уравнений Колмогорова 

### a) установившиеся вероятности состояний

```python
def findX(L, size):
    M = L.copy().transpose()
    for i in range(size):
        M[i][i] = 0
        for j in range(size):
            M[i][i] += L[i][j]
        M[i][i] *= -1

    # print("\n")
    # print(M)

    M_ = M.copy()
    for i in range(size):
        M_[m+n][i] = 1
    # print("\n")
    # print(M_)

    B = np.zeros((size, 1))
    B[m + n][0] = 1
    # print("\n")
    # print(B)
    X = np.linalg.inv(M_)@B
    return X

print(f'Установившиеся вероятности состояний:')

X = findX(L, size)
print(X)
```

### b) Найдите вероятность отказа в обслуживании

```python
def find_p_ref(X, m, n):
    return X[m+n]

p_ref = find_p_ref(X, m, n)[0]
print(f'Вероятность отказа: {p_ref}')
```

### c) Найдите относительную и абсолютную интенсивность обслуживания.

```python
def findQ(p_ref):
    return 1 - p_ref

def findA(q, la):
    return q * la

q = findQ(p_ref)
print(f'Относительная пропускная способность: {q}')

A = findA(q, la)
print(f'Абсолютная пропускная способность: {A}')
```

### d) Найдите среднюю длину в очереди.

```python
def find_L_queue(X, m, n):
    L_queue = 0
    for i in range(1, n+1):
        L_queue += i * X[m + i]
    return L_queue
    
L_queue = find_L_queue(X, m, n)   
print(f'Средняя длина очереди: {L_queue}')
```

### e) Найдите среднее время в очереди.

```python
def find_T_queue(X, n, m, ny):
    T_queue = 0
    for i in range(0,n):
        T_queue += (i + 1) / (m * ny) * X[m + i]
    return T_queue

T_queue =  find_T_queue(X, n, m, ny) 
print(f'Cреднее время в очереди: {T_queue}')
```

### f) Найдите среднее число занятых каналов.

```python
def find_N_canal(X, m, n):
    N_c1=0
    for i in range(1, m+1):
        N_c1 += i * X[i]

    N_c2=0
    for i in range(m+1,m+n):
        N_c2 += m * X[i]

    return N_c1 + N_c2

N_canal = find_N_canal(X, m, n)
print(f'Cреднее число занятых каналов: {N_canal}')
```

### g) Вероятность того, что поступающая заявка не будет ждать в очереди

```python
def p_not_wait(m):
    return sum(X[0:m])

print(f' Вероятность того, что поступающая заявка не будет ждать в очереди: {p_not_wait(m)}')
```
